```python
start_time = time.time()
annotators_user_data=[]
annotation_times=[]
work_package_size=[]
highly_disagree_responses=0
cannot_solve_response=[]
corrupt_data_response=[]
for project_id,value in anonymized_test_data['results']['root_node']['results'].items():
    positive_responses=[]
    negative_responses=[]
    for annotation_result in value['results']:
        work_package_size.append(annotation_result['workpackage_total_size'])
        annotators_user_data.append(annotation_result['user']['vendor_user_id'])
        annotation_times.append(annotation_result['task_output']['duration_ms'])
        if annotation_result['task_output']['answer']=='yes':
            positive_responses.append(annotation_result['task_output']['answer'])
        elif annotation_result['task_output']['answer']=='no':
            negative_responses.append(annotation_result['task_output']['answer'])
        elif annotation_result['task_output']['cant_solve']==True:
            cannot_solve_response.append(annotation_result['task_output'])
        elif annotation_result['task_output']['corrupt_data']==True:
            corrupt_data_response.append(annotation_result['task_output'])
    if len(positive_responses)==len(negative_responses):
        highly_disagree_responses+=1
print("--- %s seconds ---" % (time.time() - start_time))
```

```python
start_time = time.time()
number_of_bad_annotators=[]
number_of_good_annotators=[]
for row in dataset.iterrows():
    reference_response=row[1]['is_bicycle']
    bad_annotations=[]
    good_annotations=[]
    for response in annotators_data.iterrows():
        annotators_data=pd.DataFrame.from_dict(row[1]['results'])
        annotators_response_data=response[1]['task_output']
        annotators_user_data=response[1]['user']
        if (annotators_response_data['answer']=='yes' and reference_response==True) or (annotators_response_data['answer']=='no' and reference_response==False):
            good_annotations.append(annotators_user_data['vendor_user_id'])
        else:
            bad_annotations.append(annotators_user_data['vendor_user_id'])
    number_of_good_annotators.append(len(np.unique(np.array(good_annotations))))
    number_of_bad_annotators.append(len(np.unique(np.array(bad_annotations))))
print("--- %s seconds ---" % (time.time() - start_time))
```
